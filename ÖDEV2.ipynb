## ***`Akıllı Gün Planlayıcı`***

```python
while True:
    try:
        print("---= Akıllı Gün Planlayıcı =---\n")

        uyku = float(input("Günlük uyku süresi (saat): "))
        ders = float(input("Günlük ders süresi (saat): "))
        serbest = float(input("Günlük serbest zaman (saat): "))


        if uyku < 0 or ders < 0 or serbest < 0:
            print("Uyarı: Süreler negatif olamaz! Lütfen tekrar giriniz.")
            continue


        toplam_sure = uyku + ders + serbest
        if toplam_sure > 24:
            print(f"Uyarı: Toplam Süre 24 Saati Geçiyor --> ({toplam_sure} Saat). Lütfen Tekrar Giriniz.")
            continue
        break

    except ValueError:
        print("Uyarı: Lütfen Sadece Sayısal Değerler Giriniz!")

uyku_durumu = ""
if uyku < 6:
    uyku_durumu = "Yetersiz Uyku"
elif 6 <= uyku <= 8:
    uyku_durumu = "Normal Uyku"
else:
    uyku_durumu = "Fazla Uyku"

calisma_durumu = ""
if ders >= 6:
    calisma_durumu = "Yoğun Çalışma Günü"
else:
    calisma_durumu = "Normal Çalışma Günü"

genel_degerlendirme = ""
if uyku < 6 and ders >= 6:
    genel_degerlendirme = "Verimsiz Gün"
else:
    genel_degerlendirme = "Dengeli Gün"

print("\n" + "-"*31)
print(f"Uyku Durumu: {uyku_durumu}")
print(f"Çalışma Durumu: {calisma_durumu}")
print(f"Genel Değerlendirme: {genel_degerlendirme}")
print("-"*(31))
```
