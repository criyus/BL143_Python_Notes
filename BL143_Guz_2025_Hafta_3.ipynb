# ***`Döngüler`***

## ***`FOR Döngüsü`***

```python
# Ekrana 10 Defa Python Öğreniyorum Yaz

for d in range(10):
  print("Python Öğreniyorum")
```

```python
# 2 ile 10 Arasında Çift Sayıları Ekrana Yazdır

for i in range(2,11,2):
  print(i)
```

```python
# 5 ile 45 Arasında Kaç Tane Sayı Vardır ? 5 ve 45 dahil

sayac = 0
for i in range(5,46):
  sayac = sayac + 1
print(sayac)
```

```python
# Fearin İçerisinde Kaç Tane Harf Vardır ?

sayac = 0

for harf in "Fear":
  print(harf)
  sayac = sayac + 1
print(sayac,"Tane Harf Vardır")
```

```python
# 7 ile 18 Arasında Sayıların Toplamını Bulunuz

toplam = 0

for i in range(7,19):
  toplam = toplam + i
print(toplam)
```

```python
# 2 ile 6 Arasında Sayıların Çarpımını Bulunuz

carpim = 1

for i in range(2,7):
  carpim = carpim * i
print(carpim)
```

```python
# Kullanıcıya 3 Hak vererek Sisteme Girmeyi Denesin

sistem_sifresi = "Çukurova Üniversitesi"

for i in range(3):
  sifre = input("Şifreyi Giriniz:")
  if sifre == sistem_sifresi:
    print("Giriş Başarılı, Sisteme Hoş Geldiniz")
  elif i == 2:
    print("Giriş Hakkınız [3] Bitti, Sistemden Atıldınız")
  else:
    print("Şifre Yanlış ,Tekrar Deneyin")
```

## ***`WHİLE Döngüsü`***

```python
# 1'den 5'e kadar sayıları ekrana yazdıran While döngüsü

sayi = 1
while sayi <= 5:
  print(sayi)
  sayi = sayi + 1
```

```python
# Kullanıcıdan doğru şifreyi alana kadar döngü

sifre = "Çü"
giris_deneme = ""

while giris_deneme != sifre:
  giris_deneme = input("Lütfen şifreyi giriniz: ")
  if giris_deneme != sifre:
    print("Yanlış şifre, tekrar deneyin.")
print("Giriş başarılı!")
```

```python
# 10'dan geriye doğru sayan ve 0'a geldiğinde 'Fırlatma!' yazan döngü

geri_sayim = 10
while geri_sayim >= 0:
  if geri_sayim == 0:
    print("Fırlatma!")
  else:
    print(geri_sayim)
  geri_sayim = geri_sayim - 1
```
