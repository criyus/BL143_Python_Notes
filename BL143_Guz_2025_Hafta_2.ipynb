## ***`input()`*** ***`Kullanıcıdan Metin Girişi Almak`***

```python
ad = input("Adınızı Giriniz :")
soyad = input("Soyadınızı Giriniz :")

print("Hoş Geldin ",ad,soyad,".")
```

***`Not :`*** ***`Kullanıcıdan İstenilen değer Integer İse int(input()) Şeklinde Kullanılır`***

```python
kenar = int(input("Karenin Bir Kenarını Giriniz: "))
karenin_alani = kenar ** 2

print("Karenin Bir Kenarı :",kenar)
print("Karenin Alanı:",karenin_alani)
```

# ***`Koşullu İfadeler`***

```python
adınız = input("Adınızı Giriniz :")
dogum_yili = int(input("Doğum Yılınızı Giriniz :"))

yas_hesapla = 2025 - dogum_yili

print("Merhaba", adınız, "Yaşınız", yas_hesapla)

if yas_hesapla >= 18:
    print("Ehliyet Alabilirsiniz")
else:
    print("Ehliyet Alamazsınız")
```

## ***`IF Kullanımı`***

```python
yasınız = int(input("Yaşınızı Giriniz :"))

if yasınız >= 18:
    print("Ehliyet Alabilirsiniz")
```

## ***`ELSE Kullanımı`***

```python
sayi = int(input("Bir Sayı Giriniz"))

if sayi % 2 == 0:
    print("Sayı Tam Bölündü")
else:
    print("Sayı Tam Bölünmedi")
```

## ***`ELİF Kullanımı`***

```python
x = int(input("Bir Sayı Giriniz"))

if x > 0:
  print("Sayı Pozitiftir")
elif x == 0:
  print("Sayı Sıfırdır")
else :
  print("Sayı Negatiftir")
```

### ***`Ortalama Hesaplama Örneği`***

```python
vize = int(input("Vize Notunuzu Giriniz :"))
final = int(input("Final Notunuzu Giriniz :"))

ortalama = vize * 0.4 + final * 0.6

if ortalama < 40 :
  print("Kaldınız")
elif 40 < ortalama < 50:
  print("Şartlı Geçtiniz")
else:
  print("Geçtiniz")
```
