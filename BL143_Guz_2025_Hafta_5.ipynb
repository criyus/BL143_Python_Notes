```python
# Fonksiyonu Tanımla

def carpım(x,y):
  print("Çarpım:",x * y)

# Fonksiyonu Çağırma

carpım(3,4)
carpım(10,11)
carpım(4,26)

carpım(x=3,y=4)
carpım(y=4,x=3)
carpım(x=3,y=4)

carpım(4,y=3)
carpım(y=3,x=4)
# carpım(x=5,27) ÇALIŞMAZ HATA VERİR
```

```python
# Küpünü Alma Fonksiyonu Tanımla

def kup(x):
  return x ** 3

sonuc = kup(4)
print("4 ün Küpü :",sonuc)
```

```python
# en Küçüğü Bulma

def en_kucuk(a,b,c):
  if a <= b and a <= c:
    return a
  if b <= a and b <= c:
    return b
  if c <= a and c <= b:
    return c

min = en_kucuk(5,2,8)
print("En Küçük Sayı:",min)
```

```python
# Örnek

yas = 19

sehir = "Mersin"

print("Ben" , sehir ,"Doğumluyum",yas ,"Yaşındayım")
```

```python
# Format Kullanarak Düzgün Yazma

print("Ben {} Doğumlumuyum {} Yaşındayım".format(sehir,yas))
print("Ben {1} Doğumlumuyum {0} Yaşındayım".format(sehir,yas))

# F String Kullanarak

print(f"Ben {sehir} Doğumlumuyum {yas} Yaşındayım")
```

```python
# Vergili Fiyat Hesaplama

def vergili_fiyat(fiyat,vergi=0.18):
  return fiyat + fiyat * vergi

# Bilgisayar
yeni_fiyat = vergili_fiyat(100_000)
print(yeni_fiyat)

# Ekmek
yeni_fiyat1 = vergili_fiyat(1000,0.08)
print(yeni_fiyat1)
```
