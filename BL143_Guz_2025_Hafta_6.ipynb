# ***`Tuple (Değiştirilemez Veri)`***

```python
info = ("criyus","hemilton","Computer Science",1)

print(info)
print(type(info))
```

```python
info[3] = 2 # çalışmaz hata verir (tuple veriler değiştirilemez)
```

## ***`Tuple İçinde İşlem Yapma :`***

```python
print(type(info[0])) # 0. index ' in türünü alır

print(info.index('Computer Science')) # hangi indexte olduğunu gösterir

print(info.count('criyus')) # kaç tane olduğunu gösterir

print(info[1:3]) # 1. indexten 3. index'e kadar yazdırır
```

```python
pixel = (255,0,0) # RGB değer : (kırmızı çıkar)
print(pixel[0]) # Kırmızı
```

```python
# ortadaki 3 elemanı alma

sayilar = (10,20,30,40,50)
print(sayilar[1:4])
```

# ***`Dictionary (Sözlük)`***

***`Not :`*** ***`STR , İNT , TUPLE değerleri anahtar olur ama LİST olamaz.`***

```python
ogrenci = {
    "ad" : "Criyus",
    "yas" : 19,
    "bolum" : "Computer Science",
    "sehir" : "Mersin",
    1453 : "İstanbul"
}

print(ogrenci)
print(type(ogrenci))
print(ogrenci["ad"])
print(ogrenci[1453])
```

```python
# 2. YOL

ogrenci1 = dict(ad = "Criyus",yas = 19,bolum = "Computer Science")
print(ogrenci1)
```

## ***`Yeni Anahtar Ekleme ve Değer Değiştirme`***

***`Not :`*** ***`Anahtar Ekleme ve Anahtar Değiştirme işlemleri aynıdır eğer anahtar zaten varsa değiştirir yoksa ekler `***

```python
# Yeni Anahtar Ekleme
ogrenci["Konum"] = "Adana"
print(ogrenci)

# Var olan değeri değiştirme
ogrenci["yas"] = 20
print(ogrenci)
```

## ***`Eleman Silme`***

```python
car = {
    "marka" : "Mercedes",
    "model" : "C63s",
    "yil" : 2023
}


# 1. YOL
del car["model"]
print(car)

# 2. YOL
car.pop("yil")
print(car)
```

### ***`Önemli :`*** ***`Anahtar Bulunamadığı durumlarda sistem hata verir ve çöker :`***

```python
print(car["yil"]) # hata verir

print(car.get("yil")) # hata vermez None
```

## ***`Sorular :`***

```python
# dict (sözlük oluştur)

örnek = {
    "ad" : "Criyus",
    "yas" : 19,
    "bolum" : "Computer Science",
    "sehir" : "Ankara",
    "ilce" : "Kadıköy"
}
```

```python
# Anahtarları Yazdır :

for i in örnek.keys():
  print(i)
```

```python
# Değerleri Yazdır :

for i in örnek.values():
  print(i)
```

```python
# Anahtar ve Değerleri Yazdır :

for key , value in örnek.items():
  print(key,value)
```

## ***`İç İçe Sözlük (Dictionary)`***

```python
pazar = {
    "meyve1" : {
        "ad": "muz",
        "fiyat" : 50
    },
    "meyve2" : {
        "ad" : "portakal",
        "fiyat" : 40
    }
}

print(pazar["meyve1"]["ad"]) # listeler içinde belirli veriyi isteme
```

```python
#pazardaki meyveleri for ile gezip yazdırma

for i, detay in pazar.items():
  print(detay["ad"])
```
