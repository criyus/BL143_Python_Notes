# ***`Aritmatik Operatörler`***

```
print(5 + 3) # Toplama
print(3 - 2) # Çıkarma
print(3 * 2) # Çarpma
print(5 / 3) # Bölme
print(5 // 3) # Tam Bölme
print(2 ** 3) # Üs Alma
print(4 % 2) # Mod Alma
```

# ***`Karşılaştırma Operatörleri`***

```
print(5 < 3) # Küçüktür
print(4 > 3) # Büyüktür
print(4 >= 3) # Büyüktür veya Eşittir
print(2 <= 3) # Küçüktür veya Eşittir
print(5 == 3) # Eşittir
print("a" != "A") # Eşit Değildir
print(400 == 4e2) # Eşittir
```

# ***`Mantık Operatörleri`***

```
print(True and False)
print("A"=="b" and "Caecus" == "CAECUS")
print(8 != 5+2 or 3**2== 9)
print(not True)
print(not False)
print(not "a" != "e")
print(not ( 3>=2 and 2 < 5**0 ))
```
