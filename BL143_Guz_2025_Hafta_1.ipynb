# ***`Karşılaştırma Operatörleri`***

```python
print(5 < 3)
print(3 < 2)
print(12 < 65)
print(13445 < 62464)
print(1 < 3)
```

## ***`Büyüktür`***

```python
print(5 > 3)
print(3 > 2)
print(12 > 65)
print(13445 > 62464)
print(1 > 3)
```

## ***`Büyüktür veya Eşittir`***

## ***`Küçüktür veya Eşittir`***

```python
print(5 <= 5)
print(3 <= 2)
print(12 <= 65)
print(13445 <= 62464)
print(1 <= 3)
```

## ***`Eşittir`***

```python
print(5 == 3)
print(4 == 4)
print(12 == 65)
print(13445 == 62464)
print(1 == 3)
print("...................")
print(400 == 4e2)
print(300 == 3e2)
print("Fear" == "fear")
```

# ***`Mantık Operatörleri`***

## ***`AND`***

```python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print(".................")
print(5 < 3 and 3 < 2)
print(3 < 2 and 1 < 3)
print(12 < 65 and 13445 < 62464)
print(1 < 3 and 134)
print(".................")
print(400 == 4e2 and 300 == 3e2)
print(300 == 3e2 and 400 == 4e2)
print(".................")
print("fear" == "fear" and "Fear" == "fear")
```

## ***`OR`***

```python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print("..............")
print(5 < 3 or 3 < 2)
print(3 < 2 or 1 < 3)
print(12 < 65 or 13445 < 62464)
print(1 < 3 or 1)
print(".............")
print(8 != 5+2 or 3**2== 9)
print(5 == 6 or 12 == 5)
print(".............")
print(400 == 4e2 or 300 == 3e2)
print(300 == 3e2 or 400 == 4e2)
print(".............")
print("fear" == "fear" or "Fear" == "fear")
```

## ***`NOT`***

```python
print(not False)
print(not True)
print(not 5 < 3)
print(not 3 < 2)
print(not 12 < 65)
print(not 13445 < 62464)
print(not 1 < 3)
print("...................")
print(not 400 == 4e2)
print(not 300 == 3e2)
print("...................")
print(not "fear" == "fear")
```
