## ***`Hesap Makinesi`***

```python
def menu():
  print(" == MENU ==")
  print("1 - Toplama İşlemi")
  print("2 - Çıkarma İşlemi")
  print("3 - Çarpma İşlemi")
  print("4 - Bölme İşlemi")
  print("5 - Mod Alma İşlemi")
  print("0 - Çıkış Yap")

def topla(a, b):
    return a + b

def cikar(a, b):
    return a - b

def carp(a, b):
    return a * b

def bol(a, b):
  if b == 0:
    return None
  return a / b

def mod(a, b):
    return a % b


while True:
  menu() #Ekranda Menü Gösterililr

  secim = input("Seçiminiz: ").strip()

  if secim not in ["0","1","2","3","4","5"]:
    print("Yanlış Seçim Yaptınız")
    continue #döngünün başına geri git

  else :


    if secim == "0":
      print("Çıkış Yaptınız")
      break

    else :
      s1 = int(input("Birinci Sayıyı Girin: "))
      s2 = int(input("İkinci Sayıyı Girin: "))

      if secim == "1": print(topla(s1,s2))
      elif secim == "2": print(cikar(s1,s2))
      elif secim == "3": print(carp(s1,s2))
      elif secim == "4": print(bol(s1,s2))
      elif secim == "5": print(mod(s1,s2))

```

## ***`Tas Kağıt Makas`***

```python
import random

secenekler = ("taş", "kağıt", "makas") # tuple tür

kullanici_scor , bilgisayar_scor = 0,0


def kazanan(k,b):
  if (k == "taş" and b == "makas") or (k == "makas" and b == "kağıt") or (k == "kağıt" and b == "taş"):
    return "Kullanıcı Kazandı!"
  if k == b:
    return "Berabere!"
  else:
    return "Bilgisayar Kazandı!"

while True:
  print("\n'taş , kağıt , makas' tan birini seçiniz")
  print("Çıkmak İçin 'q' yazınız")
  kullanici_secim = input("Seçiminiz: ").lower().strip()

  if kullanici_secim == "q":
    print("Çıkış Yaptınız")
    break

  if kullanici_secim not in secenekler:
    print("Yanlış Seçim Yaptınız")
    continue

  bilgisayar_secim = random.choice(secenekler)
  print(f"Bilgisayar Seçimi {bilgisayar_secim}")

  winner = kazanan(kullanici_secim, bilgisayar_secim)


  if winner == "Kullanıcı Kazandı!":
    print("Bu Turu Sen Kazandın")
    kullanici_scor += 1
  elif winner == "Bilgisayar Kazandı!":
    print("Bu Turu Bilgisayar Kazandın")
    bilgisayar_scor += 1
  else :
    print("Bu Tur Berabere")

  # Display current scores after each round
  print(f"Mevcut Skor: Kullanıcı: {kullanici_scor} - Bilgisayar: {bilgisayar_scor}")

  # Check for win condition (first to 3 points)
  if kullanici_scor == 3:
    print("\n--- OYUN BİTTİ ---")
    print(f"Tebrikler! Oyunu KAZANDINIZ! (Kullanıcı: {kullanici_scor} - Bilgisayar: {bilgisayar_scor})")
    break
  elif bilgisayar_scor == 3:
    print("\n--- OYUN BİTTİ ---")
    print(f"Üzgünüm, oyunu KAYBETTİNİZ. (Kullanıcı: {kullanici_scor} - Bilgisayar: {bilgisayar_scor})")
    break
```
