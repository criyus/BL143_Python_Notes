# ***`Listeler`***

```python
# Boş Liste Oluştur

liste = []
```

```python
# Liste Sonuna "Mustafa" Ekleme

liste.append("Mustafa")

print(liste)
```

```python
# 1 Nolu İndekse "Ali" Ekle

liste.insert(1, "Ali")

print(liste)
```

```python
# 1.YOL Listeden "Ali" ' yi Sil

liste.remove("Ali")

print(liste)
```

```python
# 2. YOL Listeden "Mustafa" ' yı Sil
# İndex ile silme

liste.pop(0)

print(liste)
```

```python
# Listeden Belirtilen İndex'i Yazdırma

liste1 = ["Ali","Fear","Mustafa"]

print(liste1[0]) # 0. index
print(liste1[1]) # 1. index
print(liste1[2]) # 2. index

# Tersten

print(liste1[-1]) # Tersten 1. index
print(liste1[-2]) # Tersten 2. index
print(liste1[-3]) # Tersten 3. index
```

```python
# Listeyi Sıralama

liste2 = ["Ali","Fear","Mustafa","Celil","Kadir","Veli"]

liste2.sort()

print(liste2)
```

## ***`Listelerde İşlemler`***

```python
liste3 = [12,45,75,24,36,85,84]
```

```python
# Listedeki Eleman Sayısı

print(len(liste3))
```

```python
# Listedeki En Büyük Sayı

print(max(liste3))
```

```python
# Listedeki En Küçük Sayı

print(min(liste3))
```

```python
# Listedeki Sayısal Değerleri Toplama

print(sum(liste3))
```

```python
## Örnek Bir Program Maaş Programı

maaslar = []

while True :
  maas = int(input("Bir Sonraki Kişinin Maaşını Giriniz: "))
  maaslar.append(maas)
  bitis = input("Liste Bitti mi ?")

  if bitis == "Evet" or bitis == "evet" or bitis == "e":
    break

print("Maaş Listesi : ",maaslar)

print("En Düşük Maaş : ",min(maaslar))
print("En Yüksek Maaş : ",max(maaslar))
print("Maaş Ortalaması : ",sum(maaslar)/len(maaslar))
print("Toplam Ödenen Maaş:",sum(maaslar))
```

```python
# Meyve Listesi

meyveler = ["Elma","Armut","Muz","Kiraz"]
fiyat = ["150","55","34","13"]
```

```python
# Liste Gezme

for i in meyveler:
  print(i)
```

```python
# Listeyi İndexle Gezme

for i in enumerate(meyveler):
  print(i)
```

```python
# 2 Listeyi Beraber Gezme

for meyve, fiyat in zip(meyveler,fiyat):
  print("1 kg ", meyve,"Fiyatı : ",fiyat,"TL")
```
