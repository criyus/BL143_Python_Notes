```python
import os
DOSYA_ADİ = "banka.txt"

if os.path.exists(DOSYA_ADİ):
  with open(DOSYA_ADİ, "r", encoding="utf-8") as f:
    ilk_satir = f.readlines()[0]
    bakiye = float(ilk_satir)
else:
  bakiye = 0

def menu():
  print("=== MENÜ ===")
  print("1. Bakiye sorgula")
  print("2. Para Yatır")
  print("3. Para Çek")
  print("4. Çıkış Yap")

LİMİT = 2000
cekilen_gunluk_toplam = 0

while True:
  menu()
  secim = input("Seçiminiz: ").strip()
  if secim == "4":
    with open(DOSYA_ADİ, "r", encoding="utf-8") as f:
      satirlar = f.readlines() # tüm satirları oku

    with open(DOSYA_ADİ, "w", encoding="utf-8") as f:
      f.write(f"{bakiye}\n")
      for satir in satirlar:
        f.write(satir)
    break
  if secim == "1": print(f"Bakiyeniz: {bakiye} TL'dir.")
  elif secim == "2":
    miktar = float(input("Yatırılacak miktarı giriniz: "))
    bakiye += miktar
    print(f"{miktar} TL yatırıldı. Yeni bakiye: {bakiye} TL")
    with open(DOSYA_ADİ, "a", encoding="utf-8") as f:
      f.write(f"{miktar} TL yatırıldı.\n")
  elif secim == "3":
    miktar = float(input("Çekilecek miktarı giriniz: "))
    if miktar > bakiye:
      print("Yetersiz bakiye")
    elif miktar + cekilen_gunluk_toplam > LİMİT:
      print("Günlük para çekme limitini aşamazsınız")
    else:
      bakiye -= miktar
      cekilen_gunluk_toplam += miktar
      print(f"{miktar} TL çekildi. Yeni bakiye: {bakiye} TL")
      with open(DOSYA_ADİ, "a", encoding="utf-8") as f:
        f.write(f"{miktar} TL çekildi.\n")
```
